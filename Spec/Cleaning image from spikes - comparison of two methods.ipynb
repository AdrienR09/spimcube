```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

import despike
import spimcube.functions as fct
```

# Load some data

```python
data = np.loadtxt("/Users/pelini/my_python_packages/indev/example/data_Bsweep.txt")
```

## Plot the data without processing

```python
fig = plt.figure(figsize=(14, 10))
image = plt.imshow(data, origin='lower', cmap='inferno', aspect='auto', clim=(100, 800))
plt.xlabel("CCD pixels")
plt.ylabel("B field [T]")
```

## Processing the data with ``despike.clean``

```python
fig = plt.figure(figsize=(14, 10))

data_clean = despike.clean(data)
image = plt.imshow(data_clean, origin='lower', cmap='inferno', aspect='auto', clim=(100, 800))
plt.xlabel("CCD pixels")
plt.ylabel("B field [T]")
```

## Processing the data with ``spimcube.functions.remove_spikes``

```python
fig = plt.figure(figsize=(14, 10))

data_clean = fct.remove_spikes(data)
image = plt.imshow(data_clean, origin='lower', cmap='inferno', aspect='auto', clim=(100, 800))
plt.xlabel("CCD pixels")
plt.ylabel("B field [T]")
```

# Testing rapidity of each method

```python
%timeit despike.clean(data)
```

```python
%timeit fct.remove_spikes(data)
```

### == > ``spimcube.functions.remove_spikes`` is three order of magnitude faster than ``despike.clean``
